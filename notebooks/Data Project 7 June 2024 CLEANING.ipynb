```python
import pandas as pd
data = pd.read_csv('Studio_Ghibli.csv')
```

```python
data.head()
data.info()
data.describe()
```

```python
data.dropna()
data.drop_duplicates()
```

```python
data.fillna('unknown', inplace=True)  # Replace NaN with 'unknown'
```

```python
def duration_in_minutes(duration):
    # Split the duration string by 'h' and 'm'
    hours, minutes = duration.split('h ')
    # Convert hours and minutes to integers and calculate total minutes
    total_minutes = int(hours) * 60 + int(minutes[:-1])  # Remove 'm' from minutes and convert to integer
    return total_minutes

# Apply the function to create a new column 'Duration_Minutes'
data['Duration in Minutes'] = data['Duration'].apply(duration_in_minutes)
```
